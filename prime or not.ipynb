TAKE A INTEGER FIND PRIME OR NOT


```
num=int(input("enter the number:"))
if num>1:
  for i in range(2,num):
    if (num%i)==0:
       print(num,"is a not prime number ")
       break
  else:
    print(num,"is a prime number")
else:
  print(num,"is not a prime number")    
```
